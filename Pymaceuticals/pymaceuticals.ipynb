### Analysis

Here you'll find a top-level summary of findings from the company's most recent round of SCC tumor treatment testing. After cleaning the dataset and verifying the quality of data, a summary table on the effectiveness of each regimen on tumor size was produced along with graphs visualizing different dimensions of the mice and drugs. A cursory overview of the descriptive statistics led us to identify which regimens seemed to be the most effective: Capomulin, Ramicane, Infubinol, and Ceftamin. An indepth analysis of these regimens' effects on mice produced the following findings:

- Among the 4 drugs mentioned, Capomulin and Ramicane clearly had a more significant impact on reducing tumor volume size, on average.
- The consistency of the two drugs' effectiveness is further corroborated by the fact that both drugs were observed during more timepoints and maintained lower variance in final tumor size compared to the Infubinol and Ceftamin regimens
- Further analysis on the Capomulin regimen also yielded a key insight: a strongly correlated relationship between mouse weight and average tumor size across all timepoints.
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Remove quotations from Mouse Metadata dataframe
mouse_metadata["Mouse ID"] = mouse_metadata["Mouse ID"].apply(lambda x: x.replace('"', ''))
mouse_metadata["Drug Regimen"] = mouse_metadata["Drug Regimen"].apply(lambda x: x.replace('"', ''))
mouse_metadata["Sex"] = mouse_metadata["Sex"].apply(lambda x: x.replace('"', ''))

# Combine the data into a single DataFrame
complete_data = pd.merge(study_results, mouse_metadata, how="left", on=["Mouse ID", "Mouse ID"])

# Display the data table for preview
complete_data.head()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume.
regimen_mean = complete_data.groupby("Drug Regimen")['Tumor Volume (mm3)'].mean()
regimen_med = complete_data.groupby("Drug Regimen")['Tumor Volume (mm3)'].median()
regimen_var = complete_data.groupby("Drug Regimen")['Tumor Volume (mm3)'].var()
regimen_std = complete_data.groupby("Drug Regimen")['Tumor Volume (mm3)'].std()
regimen_sem = complete_data.groupby("Drug Regimen")['Tumor Volume (mm3)'].sem()
# Assemble the resulting series into a single summary DataFrame.
regimen_summary = {
    "Mean Tumor Volume": regimen_mean,
    "Median Tumor Volume": regimen_med,
    "Tumor Volume Variance": regimen_var,
    "Tumor Volume Std. Dev.": regimen_std,
    "Tumor Volume Std. Err.": regimen_sem
            }

regimen_summary = pd.DataFrame(regimen_summary, columns=["Mean Tumor Volume", "Median Tumor Volume",
                                                               "Tumor Volume Variance", "Tumor Volume Std. Dev.",
                                                               "Tumor Volume Std. Err."])
regimen_summary
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
max_tp=complete_data.groupby("Mouse ID")["Timepoint"].max()
max_tp=max_tp.reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merge_df=pd.merge(max_tp, complete_data, on=['Mouse ID', 'Timepoint'], how='left')

merge_df.head(10)
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatment=['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol=[]

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for drug in treatment:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    last_tumor_vol = merge_df.loc[merge_df['Drug Regimen'] == drug, 'Tumor Volume (mm3)']

    # add subset 
    tumor_vol.append(last_tumor_vol)
    
    # Determine outliers using upper and lower bounds
    quartiles = last_tumor_vol.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq  
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)

    outliers = last_tumor_vol.loc[(last_tumor_vol < lower_bound) | (last_tumor_vol > upper_bound)]

    print(f"{drug}'s potential outliers: {outliers}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
point = dict(markersize=12, markerfacecolor='red')
plt.boxplot(tumor_vol, labels=treatment, flierprops=point)
plt.ylabel('Final Tumor Volume (mm3)')
plt.show()
```
